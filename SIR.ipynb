```python
prev0 = .05
n = 1000

S = n * (1-prev0)
I = n * prev0
R = 0

a = .4
b = .1

dS = -a*S*I/n
dR = b*I
dI = -dS-dR

dSvec = []
dRvec = []
dIvec = []

r0 = -dS/dR

while dR > 1:
    dS = -a*S*I/n
    dR = b*I
    dI = -dS-dR
    S += dS
    I += dI
    R += dR
    dSvec.append(dS)
    dIvec.append(dI)
    dRvec.append(dR)

# import matplotlib.pyplot as plt 
plt.figure(figsize=(15,5))
plt.plot(dSvec)
plt.plot(dIvec)
plt.plot(dRvec)

print(r0)
```
